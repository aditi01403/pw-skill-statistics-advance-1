
1. What is a random variable in probability theory?

A random variable is a numerical outcome of a random experiment. It assigns a numerical value to each possible outcome in a probability space. There are two types:

Discrete random variables (countable outcomes, like rolling a die)

Continuous random variables (uncountable outcomes, like measuring weight)



---

2. What are the types of random variables?

Random variables are classified into two types:

Discrete Random Variable: Takes countable values (e.g., number of heads in coin toss).

Continuous Random Variable: Takes uncountable values within an interval (e.g., height of students).



---

3. What is the difference between discrete and continuous distributions?

Discrete Distribution: Represents probability for discrete random variables (e.g., Binomial, Poisson).

Continuous Distribution: Represents probability for continuous variables using probability density functions (e.g., Normal, Exponential).



---

4. What are probability distribution functions (PDF)?

A Probability Distribution Function (PDF) describes how probabilities are distributed over values of a random variable.

For discrete variables, the Probability Mass Function (PMF) is used.

For continuous variables, the Probability Density Function (PDF) is used.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

The Central Limit Theorem (CLT) states that the sampling distribution of the mean approaches a normal distribution as the sample size increases, regardless of the population's original distribution.

Helps in hypothesis testing and confidence intervals.



---

17. What are point estimates and interval estimates in statistics?

Point Estimate: Single value estimate (e.g., sample mean).

Interval Estimate: A range of values (e.g., confidence intervals).
